```python
# Se carga una segunda base de datos para hacer merge desde Python

calendardf2 = pd.read_csv('calendardf2.csv')
```

```python
calendardf2.columns
```

```python
combinadodf = pd.merge(calendardf2, listingsdf, on='listing_id')
```

```python
combinadodf.columns
```

```python
combinadodf['cancellation_policy']
```

```python
combinadodf = combinadodf[['listing_id', 'adjusted_price', 'available', 'neighbourhood', 'property_type', 'amenities', 'number_of_reviews', 'review_scores_rating', 'amenities', 'cancellation_policy']]
```

```python
combinadodf
```

```python
# Se supone que la disponibilidad f (falso) es la cantidad de reservaciones, porque es falso que esté disponible por lo que fue reservado

# Se filtran las filas para conservar solamente las que den f en la columna available

filas_f = combinadodf[combinadodf['available'] == 'f']

# Se agrupa por 'listing_id' y se calcula la suma de 'adjusted_price' por cada 'listing__id' para jerarquizar por cantidad de ventas

precio_sum_listed = filas_f.groupby('listing_id')['adjusted_price'].sum()

# Se ordena en orden descendente

precio_sum_listed_sorted = precio_sum_listed.sort_values(ascending=False)

precio_sum_listed_sorted
```

```python
# Paso 1: Eliminar duplicados
combinadodf = combinadodf.drop_duplicates(subset='listing_id', keep='first')

# Paso 2: Unir DataFrames
combinadodf2 = combinadodf.merge(precio_sum_listed_sorted, on='listing_id', how='left', suffixes=('_original', '_sum'))
```

```python
# Se vuelven a filtrar las filas para solo quedarnos con las 'f' en 'available'combinadodf2

combinadodf2 = combinadodf2[combinadodf2['available'] == 'f']
```

```python
# Se ordena de mayor a menor en cuanto a ventas sumadas en la columna'adjusted_price_sum', resultado de la suma de todos los días en 'f' por propiedad o 'listing_id'

combinadodf2 = combinadodf2.sort_values(by='adjusted_price_sum', ascending=False)
```

```python
combinadodf2
```

```python
# Se definen rangos de precios para entender cuáles son los 'listing_id' que mejor rendimiento tuvieron

rangos = [0, 1000000, 2000000, 2800000, 3647810, float('inf')]  # Puedes ajustar estos límites según tus necesidades


# Se definen las etiquetas para los rangos

etiquetas = ['1000000', '2000000', '2800000', '3647810', '3647810+']

# Se crea una nueva columna 'precio_rango' que asigna cada valor de adjusted_price a un rango

combinadodf2['precio_rango'] = pd.cut(combinadodf2['adjusted_price_sum'], bins=rangos, labels=etiquetas, right=False)

# Ahora, combinadodf contendrá la nueva columna 'precio_rango' con los rangos asignados

# Mostrar el DataFrame con las nuevas asignaciones

combinadodf2[['listing_id', 'adjusted_price_sum', 'precio_rango']]
```

```python
# Se cuenta la frecuencia de cada rango
conteo_por_rango = combinadodf2['precio_rango'].value_counts()

# Se calcula el porcentaje respecto al total que representa cada rango
porcentaje = (conteo_por_rango / len(combinadodf2)) * 100

# Se crea un DataFrame para mostrar los resultados
resultados = pd.DataFrame({'Frecuencia': conteo_por_rango, 'Porcentaje': porcentaje})

resultados
```

```python
# Se establece que solamente 282 propiedades generan cantidades millonarias al año, por lo que esas son el objeto de estudio de este proyecto
```

```python
# Filtrar las filas en el rango deseado

propiedades_top = combinadodf2[(combinadodf2['precio_rango'] == '2800000') | (combinadodf2['precio_rango'] == '3647810')]

propiedades_top
```

```python
# Antes de utilizar los datos de las mejores propiedades, hacemos una base de datos dummy para graficarla en Power BI

combinadodf2_dummy = combinadodf2
```

```python
# Se crean columnas dummy para los rangos de precios
dummy_columns_precio = pd.get_dummies(combinadodf2_dummy['precio_rango'])
dummy_columns_precio.columns = ['precio_' + col for col in dummy_columns_precio.columns]

# Crear columnas dummy para los rangos de porcentaje
dummy_columns_porcentaje = pd.get_dummies(combinadodf2_dummy['precio_rango'])
dummy_columns_porcentaje.columns = ['porcentaje_' + col for col in dummy_columns_porcentaje.columns]

# Concatenar las columnas dummy con el DataFrame original
combinadodf2_dummy = pd.concat([combinadodf2_dummy, dummy_columns_precio], axis=1)

# Mostrar el DataFrame resultante
combinadodf2_dummy
```

```python
# Reemplazar True por 1 y False por 0
combinadodf2_dummy[['precio_1000000', 'precio_2000000', 'precio_2800000', 'precio_3647810', 'precio_3647810+']] = combinadodf2_dummy[['precio_1000000', 'precio_2000000', 'precio_2800000', 'precio_3647810', 'precio_3647810+']].replace({True: 1, False: 0})
combinadodf2_dummy
```

```python
combinadodf2_dummy = combinadodf2_dummy[['listing_id', 'adjusted_price_original', 'available', 'adjusted_price_sum', 'precio_rango', 'precio_1000000', 'precio_2000000', 'precio_2800000', 'precio_3647810']]
```
