```python
# Importar librerías usadas

import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set()
```

```python
# Cargar DataFrame calendar y visualizar primeras filas para conocer disposición

calendardf = pd.read_csv('calendar.csv')
calendardf.head()
```

```python
# Revisar más filas para conocer su comportamiento, sobre todo en 'listing_id' y columnas de precios

calendardf.head(50)
```

```python
# Se comprende que no se usarán ciertas columnas, se descartan para que el DataFrame no sea tan pesado, 
# permanecen 'listing_id, 'date', 'available' y 'adjusted_price', esta última representando el precio que se pagó por la noche o la residencia

calendardf = calendardf[['listing_id', 'date', 'available', 'adjusted_price']]
```

```python
# Se limpia la columna 'adjusted_price' para convertirla en valores numéricos y poder realizar operaciones de agregación

calendardf['adjusted_price'] = calendardf['adjusted_price'].str[:-3]
calendardf['adjusted_price'] = calendardf['adjusted_price'].str.replace(r'[$,]', '', regex=True)
calendardf['adjusted_price'] = pd.to_numeric(calendardf['adjusted_price'])

# Se verifica el tipo de dato de la primera entrada en la columna 'adjusted_price'

print(type(calendardf['adjusted_price'][1]))

# Se muestran las primeras filas del DataFrame

calendardf.head()
```

```python
# Se continúa con la limpieza para quitarle el 0 y el . de al final

calendardf['adjusted_price'] = calendardf['adjusted_price'].astype(str)
calendardf['adjusted_price'] = calendardf['adjusted_price'].str[:-2]
calendardf['adjusted_price'] = pd.to_numeric(calendardf['adjusted_price'])
```

```python
cal_NO = calendardf

def extract_first_four_digits(num):
    price_str = str(num)
    numerical_str = ''.join(str(ord(char)) if not char.isdigit() else char for char in price_str)  # Convert non-digits to numerical representation
    digits = ''.join(filter(str.isdigit, numerical_str))  # Extract only digits from the string
    if digits:
        result = digits[:4]
        return int(result)
    else:
        return None  # or whatever you want to return for non-numeric or empty values

# Apply the function to the 'adjusted_price' column
cal_NO['adjusted_price'] = cal_NO['adjusted_price'].apply(extract_first_four_digits)

# Now, calendardf['price'] contains the first four digits of each element

print(cal_NO['adjusted_price'])

```

```python
# Assuming 'calendardf' is your DataFrame
# Count 'f' values for each listing_id
availability_counts = calendardf[calendardf['available'] == 'f']['listing_id'].value_counts()

# Get the top 100 listing IDs with the highest count of 'f' values
top_100_listings = availability_counts.nlargest(100)

# Extract the listing IDs
top_100_ids = top_100_listings.index.tolist()

# Print the list of top 100 IDs
print(f"The top 100 listing IDs with the highest count of 'f' values are: {', '.join(map(str, top_100_ids))}")
```

```python
# Assuming 'calendardf' is your DataFrame
# Filter the DataFrame to only include rows where 'available' is 'f'
unavailable_rows = calendardf[calendardf['available'] == 'f']

# Group by 'listing_id' and calculate the sum of 'price' for each listing_id
price_sum_by_listing = unavailable_rows.groupby('listing_id')['adjusted_price'].sum()

# Sort the Series in descending order
price_sum_by_listing_sorted = price_sum_by_listing.sort_values(ascending=False)

# Print the resulting Series
print(price_sum_by_listing_sorted)

```
